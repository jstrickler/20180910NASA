# Presidents

Read spreadsheet from the Internet, selecting the sheet name.

Use **na_values** to convert "na()" to NaN.

Set the index to the presidents' term numbers.

```python
%pylab
%matplotlib inline
import pandas as pd
df = pd.read_excel('http://qrc.depaul.edu/Excel_Files/Presidents.xlsx', sheet_name='Master',
                  na_values='NA()')
df.index = range(1,len(df)+1)
df
```

```python
%matplotlib inline
parties = df['Political Party'].value_counts()
parties
```

```python
parties.plot(kind="bar")
```

```python
occupations = df['Occupation'].value_counts()
occupations
```

```python
occupations.plot(kind="pie")
```
